#Definición de objetivo
El objetivo principal de este proyecto es identificar exoplanetas potencialmente habitables mediante el análisis de diversas características  presentes en nuestro dataframe de exoplanetas. Utilizando técnicas de filtrado y visualización de datos, nos proponemos determinar qué planetas cumplen con las condiciones necesarias para ser considerados habitables.

La identificación de exoplanetas habitables se basa en una serie de criterios específicos, como el radio planetario, la masa, la insolación recibida de su estrella anfitriona y la temperatura efectiva de la estrella. Los planetas que cumplen con estas condiciones tienen una mayor probabilidad de poseer agua líquida en su superficie, lo cual es esencial para la vida tal como la conocemos.

Los objetivos específicos del proyecto son los siguientes:


*   Filtrar Exoplanetas por Tamaño y Masa: Seleccionar planetas cuyo radio se encuentra entre 0.5 y 2.5 veces el radio de la Tierra y cuya masa está entre 0.1 y 10 veces la masa de la Tierra. Estos rangos aseguran que los planetas seleccionados sean rocosos y tengan una gravedad superficial adecuada para retener una atmósfera.

*   Evaluar la Insolación: Identificar planetas que reciben una cantidad de insolación estelar (pl_insol) entre 0.25 y 1.5 veces la insolación que recibe la Tierra del Sol. Este rango se considera adecuado para mantener temperaturas que permitan la existencia de agua líquida en la superficie del planeta.

*   Considerar la Temperatura de la Estrella: Seleccionar planetas orbitando estrellas con una temperatura efectiva (st_teff) entre 3000 K y 7000 K. Este criterio asegura que las estrellas anfitrionas no sean demasiado frías ni demasiado calientes, lo que podría afectar la habitabilidad del planeta.

*   Analizar la Distancia a la Tierra: Evaluar la distancia de los planetas habitables seleccionados a nuestro sistema solar (sy_dist) para identificar los candidatos más cercanos y, por tanto, más accesibles para futuras misiones de exploración.

*   Visualización y Comparación: Crear visualizaciones que permitan una fácil comparación de los planetas potencialmente habitables, destacando aquellos que se encuentran más cerca de la Tierra.

se puede observar que hay una correlacion directa, a mayor masa es mayor el radio del planeta. por ende lo que nos interesan son los que tengan un radio y masa relativamente pequeño. para eso empezaremos a filtrar del dataset los que cumplan las condiciones.

```python
# Calcular la matriz de correlación
corr_matrix = df[['Radio_Planeta', 'Masa_Planeta', 'Insolacion_Planeta', 'Temp_Estrella', 'Distancia_Sistema']].corr()

# Mapa de calor
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlación de las Variables', fontsize=16, fontweight='bold', color='blue')
plt.show()
```

tambien podemos ver que existe una correlacion debil entre el radio y la masa, como asi tambien entre la temperatura de la estrella con el radio del planeta, ya que a medida que el radio del planeta aumenta, la temperatura de la estrella puede tender a aumentar ligeramente

```python
# Filtra el DataFrame para obtener solo los planetas con radio entre 0.5 y 2.5
habitables_df = df[(df['Radio_Planeta'] >= 0.5) & (df['Radio_Planeta'] <= 2.5)]

# Cuenta la cantidad de filas en el DataFrame filtrado
cantidad_habitables = habitables_df.shape[0]

print("Cantidad de planetas habitables según el radio:", cantidad_habitables)
```

por lo que podemos ver tanto en el grafico como en el filtro del df, existen alrededor de 2400 planetas que cumplen la condicion del radio

```python
# Filtrar los planetas habitables según la masa
planetas_habitables_masa = habitables_df[(habitables_df['Masa_Planeta'] >= 0.09) & (habitables_df['Masa_Planeta'] <= 10)]

# Calcular la cantidad total de planetas habitables según la masa
cantidad_total_masa = len(planetas_habitables_masa)

print("La cantidad total de planetas habitables según la masa es:", cantidad_total_masa)
```

Hacemos lo mismo con la condicion de la masa , y a su vez despues filtramos por flujo estelar y temperatura

```python
# Filtrar los planetas habitables según el flujo de insolación
planetas_habitables_insol = planetas_habitables_masa[(planetas_habitables_masa['Insolacion_Planeta'] >= 0.2) & (planetas_habitables_masa['Insolacion_Planeta'] <= 1.7)]

# Calcular la cantidad total de planetas habitables según el flujo de insolación
cantidad_total_insol = len(planetas_habitables_insol)

print("La cantidad total de planetas habitables según el flujo estelar es:", cantidad_total_insol)
```

```python
# Filtrar los planetas habitables según la temperatura efectiva de la estrella
planetas_habitables_final = planetas_habitables_insol[(planetas_habitables_insol['Temp_Estrella'] >= 2900) & (planetas_habitables_insol['Temp_Estrella'] <= 7000)]

# Calcular la cantidad total de planetas habitables que cumplen con todas las condiciones
cantidad_total_final = len(planetas_habitables_final)

print("La cantidad total de planetas habitables que cumplen con todas las condiciones es:", cantidad_total_final)
```

Modelo de clasificacion usando regresion logistica

```python
# Definir los rangos de habitabilidad
rango_pl_rade = (0.5, 2.5)
rango_pl_bmasse = (0.09, 10)
rango_pl_insol = (0.2, 1.7)
rango_st_teff = (2900, 7000)

# Función para verificar la habitabilidad de un planeta
def es_habitable(planeta):
    if (rango_pl_rade[0] <= planeta['Radio_Planeta'] <= rango_pl_rade[1] and
        rango_pl_bmasse[0] <= planeta['Masa_Planeta'] <= rango_pl_bmasse[1] and
        rango_pl_insol[0] <= planeta['Insolacion_Planeta'] <= rango_pl_insol[1] and
        rango_st_teff[0] <= planeta['Temp_Estrella'] <= rango_st_teff[1]):
        return 1  # Habitabilidad
    else:
        return 0  # No habitabilidad

# Aplicar la función a cada fila del DataFrame
df['habitabilidad'] = df.apply(es_habitable, axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler

# Eliminar outliers
df_filtered = df_filtered[(df_filtered['Masa_Planeta'] < 20000)]
# Dividir los datos en características (X) y etiqueta (y)
X = df[['Radio_Planeta', 'Masa_Planeta', 'Insolacion_Planeta', 'Temp_Estrella']]
y = df['habitabilidad']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Escalar los datos
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Aplicar PCA con dos componentes principales
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train_scaled)
X_test_pca = pca.transform(X_test_scaled)
```

```python
# Crear un DataFrame con los resultados del PCA
df_pca = pd.DataFrame(data=X_train_pca, columns=['PC1', 'PC2'])
df_pca['habitabilidad'] = y_train.values

# Visualizar los resultados del PCA para las dos primeras componentes principales
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['habitabilidad'], cmap='viridis', alpha=0.6)
plt.colorbar(scatter, label='habitabilidad')
plt.title('PCA: Componentes Principales 1 vs 2')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.show()
```

```python
def train_and_evaluate_model(X_train, y_train, X_test, y_test, model, model_name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"Reporte de clasificación para {model_name}:")
    print(classification_report(y_test, y_pred,zero_division=1))

    # Matriz de confusión
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 6))
    sns.heatmap(cm, annot=True, cmap='Blues', fmt='g', cbar=False)
    plt.title(f'Matriz de Confusión - {model_name}')
    plt.xlabel('Predicción')
    plt.ylabel('Etiqueta Verdadera')
    plt.show()

# Modelo de regresión logística
model = LogisticRegression()

# Datos desbalanceados
train_and_evaluate_model(X_train_pca, y_train, X_test_pca, y_test, model, "Datos desbalanceados")

# Datos balanceados con sobremuestreo
train_and_evaluate_model(X_train_smote, y_train_smote, X_test_pca, y_test, model, "Datos balanceados con sobremuestreo")

# Datos balanceados con submuestreo
train_and_evaluate_model(X_train_under, y_train_under, X_test_pca, y_test, model, "Datos balanceados con submuestreo")
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Función para entrenar y evaluar el modelo
def train_and_evaluate_model(X_train, y_train, X_test, y_test, model, model_name):
    # Entrenar el modelo
    model.fit(X_train, y_train)

    # Hacer predicciones
    y_pred = model.predict(X_test)

    # Calcular la accuracy
    accuracy = accuracy_score(y_test, y_pred)
    print(f"La Accuracy del modelo {model_name} fue de:", accuracy * 100, "%")

    # Mostrar el reporte de clasificación
    print(f"Reporte de clasificación para {model_name}:")
    print(classification_report(y_test, y_pred))

    # Calcular la matriz de confusión
    conf_matrix = confusion_matrix(y_test, y_pred)
    print(f"Matriz de confusión para {model_name}:")
    print(conf_matrix)

    # Graficar la matriz de confusión
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No habitable', 'Habitable'], yticklabels=['No habitable', 'Habitable'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Matriz de Confusión para {model_name}')
    plt.show()

# Evaluar el mejor modelo Random Forest
train_and_evaluate_model(X_train_smote, y_train_smote, X_test_pca, y_test, best_rf, "Random Forest con SMOTE ajustado")
```
